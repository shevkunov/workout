```python
import numpy as np
import matplotlib.pyplot as plt
import math
import labwork
```

```python
N = np.array([
    [0, 884],
    [10, 907],
    [20, 807],
    [30, 740],
    [40, 666],
    [50, 606],
    [60, 526],
    [70, 467],
    [80, 412],
    [90, 375],
    [100, 343],  # !
    [95, 359],
    [85, 393],
    [75, 439],
    [65, 498],
    [55, 556],
    [45, 634],
    [35, 701],
    [25, 771],
    [15, 810],
    [10, 912],
    [5, 937]
])
```

```python
N = N[np.argsort(N[:, 0]), :]
N[:, 0] = N[:, 0] + 1
```

```python
X = 1 - np.cos(math.pi / 180 * N[:, 0])
Y = 1. / N[:, 1]
a, b, sigma_a, sigma_b = labwork.eval_mnk(X, Y)
```

```python
plt.figure(figsize=(14, 14))
grid = np.linspace(X.min() * 0.5, X.max() * 1.5, 1000)
plt.plot(grid, b * grid + a + sigma_a * 2, color="magenta")
plt.plot(grid, b * grid + a - sigma_a * 2, color="magenta")
plt.plot(X, Y)
plt.scatter(X, Y, color="red")
plt.ylim((0, Y.max() * 1.05))
plt.grid()
plt.show()
```
